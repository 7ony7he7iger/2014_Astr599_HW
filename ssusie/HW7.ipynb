```python
from astroML.datasets import fetch_rrlyrae_combined
from sklearn.cross_validation import train_test_split
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = fetch_rrlyrae_combined()

# For now, we'll only fit the first two colors
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
N_plot = 5000
plt.scatter(X[-N_plot:, 0], X[-N_plot:, 1], c=y[-N_plot:],
            edgecolors='none')
plt.xlabel('u-g color')
plt.ylabel('g-r color');
```

#Gaussian


```python
from sklearn.naive_bayes  import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

#LDA


```python
from sklearn.lda import LDA
clf = LDA()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

#SVC

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
clf.fit(X_train[::8], y_train[::8])
y_pred = clf.predict(X_test)
```

```python
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

#SVC with kernel=rbf

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
clf.fit(X_train[::2], y_train[::2])
y_pred = clf.predict(X_test)
```

```python
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

#RandomForestClassifier 

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train[::2], y_train[::2])
y_pred = clf.predict(X_test)
```

```python
print(np.sum(y_pred == y_test))
print(np.sum(y_pred != y_test))
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred,
                            target_names=['MS star', 'RR Lyrae']))
```

```python
from astroML.datasets import fetch_sdss_specgals
```

```python
data = fetch_sdss_specgals()

# put magnitudes in a matrix
X = np.vstack([data['modelMag_%s' % f] for f in 'ugriz']).T
y = data['z']

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
est = LinearRegression()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

```python
plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)
```

```python
rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

#Kneighbors

```python
from sklearn.neighbors import KNeighborsRegressor
est = KNeighborsRegressor()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)

plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)

rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

#SVR

```python
from sklearn.svm import SVR 
est = SVR()
est.fit(X_train[::3], y_train[::3])
y_pred = est.predict(X_test)

plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)

rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```

#RandomForestRegressor 

```python
from sklearn.ensemble import RandomForestClassifier
est = RandomForestClassifier() 
est.fit(X_train[::20], y_train[::20])
y_pred = est.predict(X_test)

plt.plot(y_test, y_pred, ',k')
plt.plot([0, 1], [0, 1], ':k')
plt.xlim(0, 0.6)
plt.ylim(0, 0.6)

rms = np.sqrt(np.mean((y_test - y_pred) ** 2))
print(rms)
```
