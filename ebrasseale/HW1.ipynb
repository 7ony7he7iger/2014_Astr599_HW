```python
import numpy as np
fname='/home/elizabeth/2014_fall_ASTR599/notebooks/data/asteroids5000.csv'
asteroids=np.genfromtxt(fname,delimiter=',')
```

```python
mask=np.isnan(asteroids)
missval=np.copy(asteroids[mask])
print 'The array asteroids5000.csv is missing', str(missval.size), 'data points.'
```

```python
realrows=np.array([],dtype=int)
for i in range(asteroids[:,0].size):
    data=np.copy(asteroids[i,:])
    mask=np.isnan(data)
    if data[mask].size==0:
        realrows=np.append(realrows,[i])
print 'The number of rows with no missing data is', str(realrows.size), 'out of 5000.'
```

```python
asteroids_strip=np.copy(asteroids[realrows])
print asteroids_strip.shape
```

```python
maxvals=np.max(asteroids_strip,axis=0)
maxstrs=[ "%.4f" % x for x in maxvals]
print 'Maximum values are', ', '.join(maxstrs)

minvals=np.min(asteroids_strip,axis=0)
minstrs=[ "%.4f" % x for x in minvals]
print 'Minimum values are', ', '.join(minstrs)

meanvals=np.mean(asteroids_strip,axis=0)
meanstrs=[ "%.4f" % x for x in meanvals]
print 'Mean values are', ', '.join(meanstrs)

stdvals=np.std(asteroids_strip,axis=0)
stdstrs=[ "%.4f" % x for x in stdvals]
print 'Standard deviations are', ', '.join(stdstrs)
```

```python
%matplotlib inline
```

```python
incline_dg=asteroids_strip[:,2]
incline_rad=np.radians(incline_dg)
sine_incline=np.sin(incline_rad)
a=asteroids_strip[:,0]
fig=plt.figure()                
plt.scatter(a,sine_incline)
print 'Although the spread is concentrated at the center, meteor orbits with smaller semi-major axes have a wider spread of inclinations with a high upper limit, while meteor orbits with high semi-major axis values tend to have low inclination.'
print 'However, there is a higher occurance of low inclinations at smaller orbits that increases until semi-major axis sizes reach about 3 AU. Beyond about 3.5 AU, the upper limit for inclination seems to taper off as orbit size increases.'
print 'It makes sense that orbits are most commonly uninclined, and the wider the orbit, the less possible it is for the meteor to maintain inclination, due to conservation of angular momentum.'
```

```python
H=asteroids_strip[:,3]
BV=asteroids_strip[:,5]
plt.scatter(BV,H)
med=0.775
plt.plot([med, med], [3, np.max(H)])
```

```python
fam1=BV<med
fam2=BV>=med

plt.scatter(a[fam1],sine_incline[fam1], c='r')
plt.scatter(a[fam2],sine_incline[fam2], c='y')

print 'I have chosen the colors yellow and red. The yellow meteors have, on average, tighter orbits, and the red meteors have a wider range of orbit sizes.'
```

```python
print 'My a vs sin(i) plot is very similar to Fig. 4 from Parker et al. 2008. The authors had similar findings, that meteors with smaller semi-major orbits were of a different color than those with wider orbits. What is more clear in the Parker et al. figure is that inclinations increase consistently with semi-major axis magnitude, and do not taper off beyond a values of 3.5 AU.'
```
