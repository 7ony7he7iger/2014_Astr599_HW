## 4. ASSIGNMENT: PLOTTING AND ANIMATING THE DRUMHEAD VIBRATIONAL MODES

Whenever you hit a drum, the drumhead vibrates following the two dimensional wave equation in polar coordinates,

$$
\frac{\partial u(r,\theta,t)}{\partial t} = c^2 \nabla^2 u(r,\theta,t) \\[5mm]
u(R_0,\theta,t) = 0
$$

with $R_0 = 1$ the radius of the drumhead and $u(r,\theta,t)$ the elevation of the drumhead. The solution is given in terms of a superposition of its vibrational modes:

$$
u_{n,k}(r,\theta,t) = \cos(t) \cos(n\theta) \mathrm{J}_n(\lambda_{n,k} r)
$$

where $\lambda_{n,k}$ is the $k^{th}$ positive root of $\mathrm{J}_n$ and we assumed $c=1$. Don't worry, if you don't understand the equation, in this assignment we will only plot the solutions step by step.

```python
import numpy as np
from scipy import *
from scipy.special import jn, jn_zeros

def drumhead_height(n,k,r,theta,t): 
    Unk = np.cos(t) * np.cos(n*theta) * jn(n, r * jn_zeros(n,k)[k-1])
    return Unk
    
print drumhead_height(0,2,0,0,0)
```

```python
%pylab inline
from pylab import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import matplotlib.gridspec as gridspec

def plot_Bessel(t):
    
    #polar coordintes 
    r = np.linspace(0,1,50)
    theta = np.linspace(0,2*np.pi,50)
    rm, thetam = meshgrid(r, theta)
    
    #cartesian coordinates
    x = np.empty_like(rm)
    y = np.empty_like(thetam)
    x = rm * np.cos(thetam)
    y = rm * np.sin(thetam)
    
    fig = plt.figure(figsize=(15,7))
    gs = gridspec.GridSpec(3, 3)
    #subplot1
    ax1 = fig.add_subplot(gs[:, 0:2], projection='3d')
    U23 = drumhead_height(2,3,rm,thetam,t)
    
    ax1.plot_surface(x, y, U23, rstride=2, cstride=2, alpha=0.5)
    cset = ax1.contourf(x, y, U23, zdir='z', offset=-0.5, cmap=cm.coolwarm)
    cset = ax1.contour(x, y, U23, zdir='x', offset=-1, cmap=cm.coolwarm)#
    cset = ax1.contour(x, y, U23, zdir='y', offset=1, cmap=cm.coolwarm)

    ax1.set_xlabel('X')
    ax1.set_xlim(-1, 1)
    ax1.set_ylabel('Y')
    ax1.set_ylim(-1,1)
    ax1.set_zlabel('U23')
    ax1.set_zlim(-0.5, 0.5)

    #subplot 2
    ax2 = fig.add_subplot(gs[0, -1], projection='3d')
    U13 = drumhead_height(1,3,rm,thetam,t)
    
    ax2.plot_surface(x, y, U13, rstride=4, cstride=4, alpha=0.5)
    cset = ax2.contourf(x, y, U13, zdir='z', offset=-0.5, cmap=cm.coolwarm)
    cset = ax2.contour(x, y, U13, zdir='x', offset=-1, cmap=cm.coolwarm)#
    cset = ax2.contour(x, y, U13, zdir='y', offset=1, cmap=cm.coolwarm)

    ax2.set_xlim(-1, 1)
    ax2.set_ylim(-1,1)
    ax2.set_zlabel('U13')
    ax2.set_zlim(-0.5, 0.5)
    
    # subplot 3
    ax3 = fig.add_subplot(gs[1, -1], projection='3d')
    U33 = drumhead_height(3,3,rm,thetam,t)
    
    ax3.plot_surface(x, y, U33, rstride=2, cstride=2, alpha=0.5)
    cset = ax3.contourf(x, y, U33, zdir='z', offset=-0.5, cmap=cm.coolwarm)
    cset = ax3.contour(x, y, U33, zdir='x', offset=-1, cmap=cm.coolwarm)#
    cset = ax3.contour(x, y, U33, zdir='y', offset=1, cmap=cm.coolwarm)

    ax3.set_xlim(-1, 1)
    ax3.set_ylim(-1,1)
    ax3.set_zlabel('U33')
    ax3.set_zlim(-0.5, 0.5)
    
    #sobplot 4
    ax4 = fig.add_subplot(gs[2, -1], projection='3d')
    U31 = drumhead_height(3,1,rm,thetam,t)
    
    ax4.plot_surface(x, y, U31, rstride=2, cstride=2, alpha=0.5)
    cset = ax4.contourf(x, y, U31, zdir='z', offset=-0.5, cmap=cm.coolwarm)
    cset = ax4.contour(x, y, U31, zdir='x', offset=-1, cmap=cm.coolwarm)#
    cset = ax4.contour(x, y, U31, zdir='y', offset=1, cmap=cm.coolwarm)

    ax4.set_xlim(-1, 1)
    ax4.set_ylim(-1,1)
    ax4.set_zlabel('U31')
    ax4.set_zlim(-0.5, 0.5)
```

```python
t = 3
print('Frame number:')
for i in xrange(0,t,1):
    # 3D plot parameters as function of iterator i
    # Call plotting function and save figure into file
    plot_Bessel(i)
    plt.savefig('frame{0:07d}.png'.format(i+1));
    plt.close()
    print i,
```

```python
# Import JSAnimation 
from JSAnimation import IPython_display, HTMLWriter
from matplotlib import animation
import Image
import glob

# Choose location of files to use
filenames=sorted(glob.glob('frame*.png'))

#Create figure
fig = plt.figure(figsize=(15,7))
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
im = plt.imshow(Image.open(filenames[0]))

# Create initializatio function
def init():
    im.set_data(Image.open(filenames[0]))
    return im,

# Create animation function
def animate(i):
    image=Image.open(filenames[i])
    im.set_data(image)
    return im,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=5, interval=20, blit=True)

#set embed_frames=True to embed base64-encoded frames directly in the HTML
anim.save('animation.html', writer=HTMLWriter(embed_frames=True))

# Show animation in ipython notebook
anim
```
