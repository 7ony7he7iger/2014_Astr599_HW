1) Create a function called drumhead_height that calculates u_n,k(r,θ,t). 

```python
import numpy as np
from scipy import *
from scipy.special import jn, jn_zeros

def drumhead_height(n,k,r,theta,t):
    lam = jn_zeros(n,k)[k-1]
    x = lam*r
    ans = np.cos(t)*np.cos(n*theta)*jn(n,x)
    return ans
```

2) Create a fuction called plot_Bessel that takes time t as an argument.

```python
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

def plot_Bessel(t):
    rarr = np.linspace(0,1,100)
    tharr = np.linspace(0,2*np.pi,100)
    rmesh, thmesh = np.meshgrid(rarr, tharr)
    xmesh = rmesh*np.cos(thmesh)
    ymesh = rmesh*np.sin(thmesh)
    
    gs = plt.GridSpec(3, 3, wspace=0.2, hspace=0.2)
    fig = plt.figure(figsize=(16, 6))
    ax1 = fig.add_subplot(gs[:3, :2],projection='3d')
    ax2 = fig.add_subplot(gs[0, 2],projection='3d')
    ax3 = fig.add_subplot(gs[1, 2],projection='3d')
    ax4 = fig.add_subplot(gs[2, 2],projection='3d')
    
    n = 2
    k = 3
    drummesh = drumhead_height(n,k,rmesh,thmesh,t)
    ax1.plot_surface(xmesh,ymesh,drummesh)
    
    n = 11
    k = 17
    drummesh2 = drumhead_height(n,k,rmesh,thmesh,t)
    ax2.plot_surface(xmesh,ymesh,drummesh2)
    
    n = 1
    k = 1
    drummesh3 = drumhead_height(n,k,rmesh,thmesh,t)
    ax3.plot_surface(xmesh,ymesh,drummesh3)
    
    n = 5
    k = 1
    drummesh4 = drumhead_height(n,k,rmesh,thmesh,t)
    ax4.plot_surface(xmesh,ymesh,drummesh4)
```

```python
t = linspace(0,2*np.pi,5)
```
